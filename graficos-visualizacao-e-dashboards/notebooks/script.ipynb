# Histograma

```python
# importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
```

```python
dataframe = pd.read_csv('../data/trees.csv')
dataframe.shape
```

```python
# data
dataframe.head()
```

```python
# Criação do histograma, considerando apenas o segundo atributo da base e com duas divisões (bins)
# a variavel hgt armazena as faixas de valores de height
hgt = np.histogram(dataframe.iloc[:, 1], bins=6)
hgt
```

```python
# Visualização do histograma com 6 divisões (bins)
plt.hist(dataframe.iloc[:,1], bins=6)
plt.title("Árvores")
plt.ylabel("Frequência")
plt.xlabel("Altura")
```

# Gráfico de dispersão

```python
dataframe = pd.read_csv('../data/trees.csv')
dataframe.head()
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x=dataframe.Girth, y=dataframe.Volume, color='blue', facecolor='None', marker='*')
plt.title('Árvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
# Gráfico de linhas considerando o volume e o atributo Girth sem marker
plt.plot(dataframe.Girth, dataframe.Volume)
plt.title('Árvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
# Grafico de dispersão com afastamento dos dados (jitter)
# fit_reg = linha de tendência
sns.regplot(x=dataframe.Girth, y=dataframe.Volume, data=dataframe, x_jitter=0.3, fit_reg=False)
```

# Gráficos de dispersão com seaborn

```python
# Carregamento dos dados
df_co2 = pd.read_csv('../data/co2.csv')
df_co2.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake, agrupando pelo type
sns.scatterplot(x = df_co2['conc'], y= df_co2['uptake'], hue=df_co2.Type)
```

```python
# Seleção de registros especificos da base de dados (Quebec, Mississippi)
q = df_co2.loc[df_co2['Type'] == 'Quebec']
m = df_co2.loc[df_co2['Type'] == 'Mississippi']
```

```python
# Subgráfico (1 linha e 1 coluna) mostrando graficos sobre cada região
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x= q['conc'], y= q['uptake']).set(title='Quebec')
plt.subplot(1,2,2)
sns.scatterplot(x = m['conc'], y= m['uptake']).set(title='Mississippi')
plt.tight_layout()
```

```python
# Separação entre refrigerado e não refrigerado (chilled e nonchilled)
ch = df_co2.loc[df_co2['Treatment'] == 'chilled']
nc = df_co2.loc[df_co2['Treatment'] == 'nonchilled']
```

```python
# Gráfico somente com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x= ch['conc'], y=ch['uptake']).set(title='Chilled')
plt.subplot(1,2,2)
sns.scatterplot(x= nc['conc'], y=nc['uptake']).set(title='Non chilled')
plt.tight_layout()
```

# Gráficos 3D


```python
df_orchard = pd.read_csv('../data/orchard.csv')
df_orchard
```

```python
# Criação de grafico 3D, indicando atributo projection = '3d' e passando três atributos (decrease, rowpos e colpos)
figura = plt.figure()
eixo = figura.add_subplot(1,1,1, projection='3d')
eixo.scatter(df_orchard.decrease, df_orchard.rowpos, df_orchard.colpos)
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
```
