# Histograma

```python
# importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataframe = pd.read_csv('../data/trees.csv')
dataframe.shape
```

```python
# data
dataframe.head()
```

```python
# Criação do histograma, considerando apenas o segundo atributo da base e com duas divisões (bins)
# a variavel hgt armazena as faixas de valores de height
hgt = np.histogram(dataframe.iloc[:, 1], bins=6)
hgt
```

```python
# Visualização do histograma com 6 divisões (bins)
plt.hist(dataframe.iloc[:,1], bins=6)
plt.title("Árvores")
plt.ylabel("Frequência")
plt.xlabel("Altura")
```

# Gráfico de dispersão

```python
dataframe = pd.read_csv('../data/trees.csv')
dataframe.head()
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x=dataframe.Girth, y=dataframe.Volume, color='blue', facecolor='None', marker='*')
plt.title('Árvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
# Gráfico de linhas considerando o volume e o atributo Girth sem marker
plt.plot(dataframe.Girth, dataframe.Volume)
plt.title('Árvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
# Grafico de dispersão com afastamento dos dados (jitter)
# fit_reg = linha de tendência
sns.regplot(x=dataframe.Girth, y=dataframe.Volume, data=dataframe, x_jitter=0.3, fit_reg=False)
```

# Divisão de tela

```python
# girth com volume
plt.scatter(dataframe.Girth, dataframe.Volume)
```

```python
# imprimindo juntos
# criação de figura e posicionamento dos dados

plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(dataframe.Girth, dataframe.Volume)
plt.subplot(2,2,2)
plt.scatter(dataframe.Girth, dataframe.Height)
plt.subplot(2,2,3)
plt.scatter(dataframe.Height, dataframe.Volume, marker='*')
plt.subplot(2,2,4)
plt.hist(dataframe.Volume)
```

# BOXPLOT

```python
# Gerando boxplot 
# patch_artist = True preencher showfliers outliers
plt.boxplot(dataframe.Volume, vert=False, showfliers=False, notch= True, patch_artist= True)
plt.title('Árvores')
plt.xlabel('Volume')
```

```python
# Dados por linha
plt.boxplot(dataframe)
plt.title('Árvores')
plt.xlabel('Dados')
```

```python
# Geração de 3 boxplots, cada um mostrando informações diferentes
plt.boxplot(dataframe.Volume, vert=False)
plt.boxplot(dataframe.Girth, vert=False)
plt.boxplot(dataframe.Height, vert=False)
plt.title('Árvores')
plt.xlabel('Dados')
```

# Gráfico de barras e de setores

# HISTOGRAMAS DE DENSIDADE COM SEABORN

```python
df_trees = pd.read_csv('../data/trees.csv')
df_chicken = pd.read_csv('../data/chicken.csv')
```

```python
df_trees.head()
```

```python
# Histograma com 10 divisões (bins) e com grafico de densidade(kernel density - suavisação)
# stats: count, frequency, probability, percent, density
sns.histplot(df_trees.Volume, color='blue', kde=True, stat="density", bins=10).set(title='Volume')
```

```python
df_chicken.head()
```

```python
# Criação de dataframe levando em consideração a coluna feed somando os valores de weight
agrupado =  df_chicken.groupby(['feed'])['weight'].sum()
agrupado
```

```python
# Novo agrupamento para teste de filtros do pandas
teste = df_chicken.loc[df_chicken['feed'] == 'horsebean']
teste
```

```python
# Histograma considerando apenas o valor horsebean
# apenas grafico de densidade
sns.kdeplot(df_chicken.loc[df_chicken['feed']=='horsebean'].weight).set(title="horsebean")
```

```python
# Histograma considerando apenas o valor casein
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'casein'].weight, kde=True).set(title='casein')
```

```python
# Histograma considerando apenas o valor linseed
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'linseed'].weight, kde=True).set(title='linseed')
```

```python
# Histograma considerando somente o valor meatmeal
sns.histplot(df_chicken.loc[df_chicken['feed'] == 'meatmeal'].weight, kde=True).set(title='meatmeal')
```
