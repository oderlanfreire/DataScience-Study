# Histograma

```python
# importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataframe = pd.read_csv('../data/trees.csv')
dataframe.shape
```

```python
# data
dataframe.head()
```

```python
# Criação do histograma, considerando apenas o segundo atributo da base e com duas divisões (bins)
# a variavel hgt armazena as faixas de valores de height
hgt = np.histogram(dataframe.iloc[:, 1], bins=6)
hgt
```

```python
# Visualização do histograma com 6 divisões (bins)
plt.hist(dataframe.iloc[:,1], bins=6)
plt.title("Árvores")
plt.ylabel("Frequência")
plt.xlabel("Altura")
```

# Gráficos de Densidade

```python
dataframe = pd.read_csv('../data/trees.csv')
dataframe.head()
```

```python
# Histograma com 10 divisões (bins) e somente para o primeiro atributo da base de dados
plt.hist(dataframe.iloc[:1], bins=6)
```

```python
# Histograma com a linha de distribuição de frequencia, com 6 divisões (bins)
# kde = linha de densidade
sns.histplot(dataframe.iloc[:,1], kde=False, bins=6, color='blue').set(title="Árvores") 
```

```python
sns.kdeplot(dataframe.iloc[:,1], color="blue").set(title="Árvores")
```

```python
sns.histplot(dataframe.iloc[:,1], bins=6, kde=True, color="blue").set(title="Árvores")
```

# Gráfico de dispersão

```python
dataframe = pd.read_csv('../data/trees.csv')
dataframe.head()
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x=dataframe.Girth, y=dataframe.Volume, color='blue', facecolor='None', marker='*')
plt.title('Árvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
# Gráfico de linhas considerando o volume e o atributo Girth sem marker
plt.plot(dataframe.Girth, dataframe.Volume)
plt.title('Árvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
# Grafico de dispersão com afastamento dos dados (jitter)
# fit_reg = linha de tendência
sns.regplot(x=dataframe.Girth, y=dataframe.Volume, data=dataframe, x_jitter=0.3, fit_reg=False)
```
