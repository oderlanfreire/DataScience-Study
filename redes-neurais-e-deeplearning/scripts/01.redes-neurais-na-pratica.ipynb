# Redes Neurais na pratica

```python
#imports de libs
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import numpy as np
```

```python
# Carregamento da base de dados e criação dos provedores (variaveis independentes - x) e da classe (variavel dependente - y)
df = datasets.load_iris()
previsores = df.data
classe = df.target
classe
```

```python
# Transformação da classe para o formato "dummy", pois temos uma rede neural com 3 neuronios na camada de saida
classe_dummy = to_categorical(classe)
classe_dummy
```

```python
# Divisão de dados entre treinamento e teste (30% para teste e 70% para treino)
x_train, x_test, y_train, y_test = train_test_split(previsores, classe_dummy, test_size=0.3,random_state=0)
```

```python
# Criação da estrutura da rede neural com classe sequencial (sequencia de camadas)
modelo = Sequential()
# primeira camada oculta, 5 neuronios, 4 neuronios de entrada
modelo.add(Dense(units=5, input_dim=4))
# segunda camada oculta
modelo.add(Dense(units= 4))
# terceira camada oculta
# Função softmax, por conta de um problema de classificação com mais duas classes
#(é gerado uma probabilidade em cada neurônio)
modelo.add(Dense(units=3, activation='softmax'))
```

```python
#Visualização da estrutura da rede neural
modelo.summary()
```

```python
# Configuração dos parâmetros da rede neural (adam = algoritmo para atualizar os pesos e loss = cálculo de erro)
modelo.compile(optimizer= 'adam', loss= 'categorical_crossentropy', metrics=['accuracy'])
#Treinamento, dividindo a base de treinamento em uma porção para validação (validation_data)
modelo.fit(x_train, y_train, epochs=1000, validation_data= (x_test, y_test))
```

```python
# previsões e mudar a variável para true ou false de acordo com o threshold 0.5
previsores = modelo.predict(x_test)
previsores = (previsores > 0.5)
previsores
```

```python
# Como é um problema com três saidas, precisamos buscar a posição que possui o maior valor (são retornado 3 valores)
y_test_matrix = [np.argmax(t) for t in y_test]
y_previsao_matrix = [np.argmax(t) for t in previsores]
```

```python
# geração da matriz de confusão
confusao = confusion_matrix(y_test_matrix, y_previsao_matrix)
confusao
```
